### Objectives:
- Visualize **monthly sales trends**
- Identify **top products & customers**
- Analyze **return behavior**
- Explore **customer retention & RFM**



## Database Connection

```python
from sqlalchemy import create_engine
import pandas as pd

# Connect to MySQL
engine = create_engine("mysql+mysqlconnector://root:loveuthree3@localhost/online_retail")
```

## Dataframes


```python

df_revenue = pd.read_sql("SELECT * FROM monthly_sales_trend", engine)
df_products=pd.read_sql("select * from top_10_revenue_products",engine)
df_country_revenue=pd.read_sql("select * from total_revenue_per_country",engine)
df_unique_customers_per_country=pd.read_sql("select * from unique_customers_per_country",engine)
df_most_frequent=pd.read_sql("select * from most_frequently_bought_items",engine)



```

## Monthly Sales Trend üìà

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

plt.figure(figsize=(10,5))
sns.lineplot(data=df_revenue, x='month', y='total_revenue', marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue (¬£)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Top Revenue Products

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))


sns.barplot(data=df_products, x='revenue', y='description', palette="Blues_d")


plt.title("Top Products by Revenue", fontsize=14)
plt.xlabel("Revenue")
plt.ylabel("Product Description")



plt.tight_layout()
plt.show()
```

```python



df_country_revenue.set_index('country', inplace=True)


plt.figure(figsize=(8, 14))
sns.heatmap(df_country_revenue, annot=True, fmt=".0f", cmap='viridis', linewidths=0.5)


plt.title("Revenue by Country (Heatmap)", fontsize=14)
plt.xlabel("Revenue")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

## Unique Customers Per Country

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12, 6))
sns.barplot(data=df_unique_customers_per_country,x='country',y='unique_customers', palette="Blues_d")

plt.title("Unique Customers Per Country", fontsize=14)
plt.xlabel("Country")
plt.ylabel("Customers")
plt.xticks(rotation=65,ha='right')
plt.tight_layout()
plt.show()
```

## Most Popular Items

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df_most_frequent.head(10), y='description', x='times_bought', palette="viridis")
plt.title("Top 10 Most Frequently Bought Products", fontsize=14)
plt.xlabel("Number of Invoices")
plt.ylabel("Product")
plt.tight_layout()
plt.show()
```
